```python
# Date: 23 March, 2024
# CSC461 – Assignment2 – IDS – Data-Visualization
# Badar Munir
# Sp20-Bse-068
#The pak-heritage-sites dataset visualization.

import pandas as pd
import folium

df = pd.read_csv('pak-heritage-sites.csv', header=None, names=['lat', 'lon', 'text'])

pakistan_map = folium.Map(location=[30.3753, 69.3451], zoom_start=5, tiles='OpenStreetMap')

for index, row in df.iterrows():
    folium.Marker(
        [row['lat'], row['lon']],
        popup=row['text'],
        icon=folium.Icon(color='green', icon='info-sign')
    ).add_to(pakistan_map)
pakistan_map.save('pakistan_heritage_sites_map.html')

pakistan_map

```
