```python
# Date: 23 March, 2024
# CSC461 – Assignment2 – IDS – Data-Visualization
# Badar Munir
# Sp20-Bse-068
#The world_pop question 2 dataset visualization.

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('world_pop.csv')

least_populous_2015 = df.nsmallest(10, 'year_2015')

plt.figure(figsize=(10, 6))
plt.barh(least_populous_2015['country'], least_populous_2015['year_2015'] / 1e6)
plt.xlabel('Population in 2015 (Millions)')
plt.ylabel('Countries')
plt.title('10 Least Populous Countries in 2015')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('world_pop.csv')

countries = ['Pakistan', 'India', 'United States', 'United Kingdom']
years = [str(year) for year in range(1970, 2011)]
df_years = df.set_index('country').loc[countries, [f'year_{year}' for year in years]].transpose()
df_years.plot(figsize=(14, 8))
plt.ylabel('Population')
plt.title('Population Change 1970 to 2010')
plt.legend(title='Country')
plt.xticks(np.arange(0, len(years), 5), labels=years[::5])  # Show every 5th year for clarity
plt.grid(True)
```

```python
years = [str(year) for year in range(2010, 2021)]
df_years = df.set_index('country').loc['Pakistan', [f'year_{year}' for year in years]].transpose()
df_years.plot(figsize=(10, 6), marker='o')
plt.ylabel('Population')
plt.title('Pakistan Population Growth 2010-2020')
plt.xticks(np.arange(0, len(years), 1), labels=years)  # Show each year
plt.grid(True)

plt.show()
```
