```python
# Date: 23 March, 2024
# CSC461  Assignment2  IDS  Data-Visualization
# Badar Munir
# Sp20-Bse-068
#The nuclear dataset visualization.

import pandas as pd
import folium

df = pd.read_csv('nuclear_waste_sites.csv') 

map_center = [39.50, -98.35]  # Roughly the geographic center of the US

us_map = folium.Map(location=map_center, zoom_start=4, tiles='OpenStreetMap')

for index, row in df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=folium.Popup(row['text'], parse_html=True),
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(us_map)

us_map.save('nuclear_waste_sites_map.html')  
us_map  
```
