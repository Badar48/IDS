```python
# Date: 23 March, 2024
# CSC461  Assignment2  IDS  Data-Visualization
# Badar Munir
# Sp20-Bse-068
#The world_pop dataset visualization.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('world_pop.csv')

top_10_countries_2020 = df.nlargest(10, 'year_2020')['country'].tolist()

top_10_df = df[df['country'].isin(top_10_countries_2020)]

years = [int(year.split('_')[1]) for year in top_10_df.columns[1:]]  
population_data = {}

for country in top_10_countries_2020:
    country_data = top_10_df[top_10_df['country'] == country]
    
    population_values = country_data.iloc[0, 1:].values  
    population_data[country] = population_values

# Plotting
plt.figure(figsize=(15, 10))
for country, populations in population_data.items():
    plt.plot(years, populations, label=country)

plt.title('Top 10 Most Populated Countries (1960-2020)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.xticks(years[::5], rotation=45)  
plt.legend(title='Country', title_fontsize='13', fontsize='12')
plt.grid(True)
plt.tight_layout()

plt.show()
```
